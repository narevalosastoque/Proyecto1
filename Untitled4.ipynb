```python
#Se genera el nuevo DataFrame

df = pd.read_excel('Bicicleta.xlsx')
df.head()
```

```python
#Informacin del DataFrame
df.info()
```

```python
#Se confirma si existen datos nulos en el Dataframe.
df.isnull().sum()
```

```python
#Se generra un nuevo DataFrame numrica.
df_num = df.loc[:,['BikeBuyer','AvgMonthSpend','YearlyIncome','NumberCarsOwned','Age','NumberChildrenAtHome','TotalChildren']]
df_num.head()
```

```python
#Se va a convertir la variable BikeBuyer en Cualitativa
df_num['BikeBuyer'] = df['BikeBuyer'].apply(str)
df_num.head()
```

```python
df_num.info()
```

```python
df_num.describe().T
```

```python
df_pal = df.loc[:,['Gender','Age_Bin','Occupation','MaritalStatus']]
df_pal.head()
```

```python
#Importar librerias de grficos
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import missingno as msno
```

```python
#Grfica de Barras
plt.figure(figsize=(7,5))
sns.countplot(df_pal['Age_Bin'], order=df_pal['Age_Bin'].value_counts().index)
plt.show()
```

```python
#Unimos las variables Age_Bin y Bike Buyer en un nuevo Data Frame
df_BikeBuyer = pd.crosstab(df_pal['Age_Bin'], df_num['BikeBuyer'])
df_BikeBuyer
```

```python
#Generamos una nueva grfica con esas variables.
df_BikeBuyer.plot.bar()
plt.show()
```

```python
#Se geenra un nuevo DataFrame para calcular la frecuencia de la variable gnero.
nivel_frec = (df_pal.
              groupby('Gender')
              .agg(frequency=('Gender','count')))
nivel_frec
```

```python
#Grfica de la variable Gnero.
labels = ['Female','Male']

nivel_frec.frequency.plot.pie(figsize=(6,6),
                              autopct='%1.1f%%',
                              labels=labels
                              );
```

```python
#Grfica de la Variable ocupacin.
plt.figure(figsize=(7,5))
sns.countplot(df_pal['Occupation'], order=df_pal['Occupation'].value_counts().index)
plt.show()
```

```python
#Se genera una tabla de frecuencias para la variable estado civil
nivel_frec = (df_pal.
              groupby('MaritalStatus')
              .agg(frequency=('MaritalStatus','count')))
nivel_frec
```

```python
#Se hace la grfica de pastel para la variable de estado civil
labels = ['Married','Single']

nivel_frec.frequency.plot.pie(figsize=(6,6),
                              autopct='%1.1f%%',
                              labels=labels
                              );
```

```python
#Se grafica las variables de ingresos Anuales con rango de edad en un Boxplot
fig, ax = plt.subplots(figsize=(8,8))
sns.boxplot(x = 'Age_Bin', y = 'YearlyIncome', data = df, ax = ax)
ax.set_title('Ingreso Anual por Rango de Edad');
```

```python
#Se grafica un diagrama de Correlacin entre las variables el gasto promedio mensual con el nmero de hijos en casa
sns.jointplot(x=df_num['YearlyIncome'], y=df_num['AvgMonthSpend'])
```

```python
a= sns.jointplot(x=df_num['YearlyIncome'], y=df_num['AvgMonthSpend'])
sns.regplot(df_num['YearlyIncome'],df_num['AvgMonthSpend'], ax=a.ax_joint, scatter=False, color='red');
```

```python
#Mapa de Calor 
a= sns.jointplot(x='YearlyIncome', y='AvgMonthSpend', data=
                 df_num, kind='hex', cmap='hot', size= 6)
sns.regplot(df_num['YearlyIncome'],df_num['AvgMonthSpend'], ax=a.ax_joint, scatter=False, color='yellow')
```

```python
#Creacin de un DataFrame llamado df2

col_1 = ['AvgMonthSpend','YearlyIncome','NumberCarsOwned','Age','NumberChildrenAtHome','TotalChildren']
df2 = df[col_1]
df2.head()
```

```python
#Grafica de grilla de correlacin
correlacion = df2.corr()
mask = np.triu(np.ones_like(correlacion, dtype=bool))
f, ax = plt.subplots(figsize=(10,10))
sns.heatmap(correlacion, mask= mask, cmap='Blues',vmax=1, center=0,
            square=True, linewidths=.1, annot=True)
```

```python
#Grafico de puntos y curvas
plt.figure(figsize=(6,4))
sns.pairplot(df2, diag_kind='kde');
```
